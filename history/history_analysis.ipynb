### Meaning of fields
* `tMyr` -- This is time in units of Myr. The same with what you can get from the corresponding metalicity pickle file.
* `v_2p` -- Mass-weighted velocity dispersion (you can condier it as a turbulent velocity) of the "warm gas" (which I define with temperature cut; $5\times10^3K<T<2\times10^4K$)
$$ v \equiv \left(\frac{\int \rho v^2 dV}{\int \rho dV}\right)^{1/2}$$
* `H_2p` -- Scale height of the "warm gas". You may define turbulent diffusion coefficient using $\kappa_{\rm turb}\sim vH$
* `v_u` and `H_u` -- same properties for the "colder gas" ($T<5\times10^3K$), which may be more directly related to star formation
* `sfr10` and `sfr40` -- star formation rate surface density for last 10Myr and 40Myr, respectively, in units of ${\rm M_\odot/pc^2/Myr}$ 
* `snr10` and `snr40` -- supernova rate surface density for last 10Myr and 40Myr, respectively, in units of ${\rm \#/pc^2/Myr}$ 

```python
plt.plot(hstzp['tMyr'],hstzp['v_u']*hstzp['H_u'],label='cold')
plt.plot(hstzp['tMyr'],hstzp['v_w']*hstzp['H_w'],label='warm')
plt.legend()
plt.yscale('log')
plt.xlabel('t [Myr]')
plt.ylabel(r'$\kappa_{\rm turb} [{\rm km/s\cdot pc}]$')
```

```python
plt.plot(hstzp['tMyr'],hstzp['sfr10'])
plt.plot(hstzp['tMyr'],hstzp['sfr40'])
plt.ylabel(r'$\Sigma_{\rm SFR} [{\rm M_\odot/pc^2/Myr}]$')
plt.xlabel('t [Myr]')
```

```python

plt.plot(hstzp['tMyr'],hstzp['snr10'])
plt.plot(hstzp['tMyr'],hstzp['snr40'])
plt.ylim(5.e-6,5.e-4)
plt.yscale('log')
plt.ylabel(r'$N_{\rm SNR} [{\rm \#/pc^2/Myr}]$')
plt.xlabel('t [Myr]')
```
