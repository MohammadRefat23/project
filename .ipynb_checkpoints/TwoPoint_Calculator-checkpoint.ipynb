```python
### Importing matplot and packages

%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import cPickle as p
```

```python
### Load the metal data file ###

Z = p.load(open('data/R8_8pc_metal.0100.scal0.p','rb'))
```

```python
### Loading the data on the plot ###

Zxy_data = Z['z']["data"]
Zxy_bounds = Z['z']["bounds"]
Nx,Ny = Zxy_data.shape
```

```python
### Creating the plot ###

plt.imshow(Zxy_data, extent = Zxy_bounds)
```

```python
### Defining the radius ###

radius_array = np.arange(11)
phi = np.linspace(0.,2*np.pi,10.)

### Defining the Simulation bounds ###

Sim_range = np.linspace(Zxy_bounds[0],Zxy_bounds[1],Zxy_data.shape[0])
Sim_range_size = Sim_range.size
Sim_space = Sim_range[1]-Sim_range[0]
```

```python
### Looping over all of the radii to create the scatter plot ###

for i in range(len(radius_array)):
    radius = radius_array[i]
    dx = radius * np.cos(phi)
    dy = radius * np.sin(phi)
```
