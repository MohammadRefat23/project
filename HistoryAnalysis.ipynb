```python
### Importing packages 

import numpy as np
import matplotlib.pyplot as plt
import cPickle as p
import glob
import os
```

```python
### Setting Image parameters

plt.rcParams['figure.figsize']=(15,8)
plt.rcParams['font.size']=30
```

```python
### Setting Time, Lambda and Sigma arrays to their own variables

Time = np.loadtxt("TimeArray.txt")
Lambda = np.loadtxt("LambdaArray.txt")
sigma = np.loadtxt("SigmaArray.txt")
```

```python
### Reading in the history table

hstzp=p.load(open("./history/hst/R8_8pc_metal.hst_zp.p"))
```

```python
### Lambda vs. Time

plt.plot(Time,Lambda,label= r"$ \lambda $")
plt.xlim(200,500)
plt.xlabel("Time [Myr]")
plt.ylabel(r"$ \lambda $")
plt.legend()

###plt.savefig("./LambdaPlot.png")
```

```python
### Lambda & Sigma

plt.plot(Time,Lambda/Lambda.mean(),label= r"$ \lambda/ \langle \lambda \rangle $")
plt.plot(Time,sigma/sigma.mean(), label = r"$  \sigma /\langle \sigma \rangle $")
plt.xlim(200,500)
plt.xlabel("Time [Myr]")
plt.ylabel(r"$ \alpha/ \langle \alpha \rangle $")
plt.legend()

###plt.savefig("./LambdaSigmaPlot.png")
```

```python
### Lambda & SFR10

plt.plot(hstzp["tMyr"],hstzp["sfr10"]/hstzp["sfr10"].mean(),label="SFR10")
plt.plot(hstzp["tMyr"],Lambda/Lambda.mean(),label=r"$ \lambda/ \langle \lambda \rangle $")
plt.legend()
plt.xlim(200,500)
plt.xlabel('t [Myr]')
plt.ylabel(r"$ \alpha/ \langle \alpha \rangle $")

###plt.savefig("./LambdaSFR10.png")
```

```python
### Lambda, Sigma & SFR40

plt.plot(hstzp["tMyr"],hstzp["sfr40"]/hstzp["sfr40"].mean(),label="SFR40")
plt.plot(hstzp["tMyr"],Lambda/Lambda.mean(),label="Lambda")
plt.plot(hstzp["tMyr"],sigma/sigma.mean(),label="Sigma")
plt.legend()
plt.xlabel('t [Myr]')
```

```python
### Lambda, Sigma & SNR10
plt.plot(hstzp["tMyr"],Lambda/Lambda.mean(),label=r"$ \lambda/ \langle \lambda \rangle $")


plt.plot(hstzp["tMyr"],hstzp["snr10"]/hstzp["snr10"].mean(),label="SNR10")
plt.legend()
plt.xlim(200,500)
plt.ylabel(r"$ \alpha/ \langle \alpha \rangle $")
plt.xlabel('t [Myr]')

###plt.savefig("./LambdaSNR10.png")
```

```python
### Lambda, Sigma & SNR40

plt.plot(hstzp["tMyr"],hstzp["snr40"]/hstzp["snr40"].mean(),label="SNR40")
plt.plot(hstzp["tMyr"],Lambda/Lambda.mean(),label="Lambda")
plt.plot(hstzp["tMyr"],sigma/sigma.mean(),label="Sigma")
plt.legend()
plt.xlabel('t [Myr]')
```
