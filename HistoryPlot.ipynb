```python
### Importing packages 

import numpy as np
import matplotlib.pyplot as plt
import cPickle as p
import glob
import os
```

```python
### Bringing the table in

hstzp_filename='./hst/R8_8pc_metal.hst_reduced.p'
hstzp=p.load(open(hstzp_filename))
```

```python
### The time and Lambda plots

LambdaArray = np.loadtxt("LambdaArray.txt")
TimeArray = np.loadtxt("TimeArray.txt")
```

```python
### Plotting Lambda vs Time

plt.plot(TimeArray,LambdaArray)
```

```python
### Plotting Lambda, Time, V and H

plt.plot(TimeArray,hstzp['v_u']*hstzp['H_u'])
plt.plot(TimeArray,hstzp['v_w']*hstzp['H_w'])
```

```python
### Plotting Lambda, Time and SFR

plt.plot(TimeArray,hstzp['sfr10'])
plt.plot(TimeArray,hstzp['sfr40'])
```

```python
### Plotting Lambda, Time, SFR and V

plt.plot(TimeArray,hstzp['sfr10'])
plt.plot(TimeArray,hstzp['sfr40'])
plt.plot(TimeArray,hstzp['v_u']*hstzp['H_u'])
plt.yscale('log')
```
